```python
#imports for the project

import numpy as np
import matplotlib.pyplot as plt



%matplotlib inline
```

```python
#Ordinary least squares linear regression practice

#generate some input data
X = np.arange(-5,5,0.1)
X
```

```python
#Let Y = f(x) = 3 + 0.5n - n^2 + 0.15n^3 be the function we are trying to 'find'.  It is the 'solution'
#compute f(x) for inputs...

Y = 3 + 0.5 * X - X**2 + 0.15 * X**3
Y
```

```python
#plot the function
plt.plot(X,Y,'b*')
plt.show
```

```python
#generate some random inputs,
trainX = np.random.random(20) * 10 - 5
trainX
```

```python
plt.hist(trainX)
```

```python
#compute & display Y with and without noise + the true function we're trying to discover
    
trainY = 3 + 0.5 * trainX - trainX**2 + 0.15 * trainX**3
plt.plot(trainX, trainY, '.k')
plt.title('Random Data without noise')
plt.show()

trainY = trainY + (np.random.randn(trainY.size)*3)

plt.plot(trainX, trainY, '.k')
plt.title('Random Data with noise')
plt.show()

plt.plot(trainX, trainY, '.k', label='Solution Function')
plt.plot(X, Y, 'b-', label='Model')
plt.legend(loc=4)
plt.show()
```

```python
#now generate phi using polynomials of X and perform OLS regression
Phi = np.array([trainX ** p for p in range(6)]).T
w = np.linalg.inv(Phi.T @ Phi) @ Phi.T @ trainY
w
```

```python
#plot results
Yhat = np.array([X ** p for p in range(6)]).T @ w
plt.plot(X, Yhat, 'r-', X, Y, 'b-', trainX, trainY, 'k.')
plt.show()
```
